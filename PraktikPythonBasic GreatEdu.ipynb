**NOMOR 3**

```python
#4 Variable dengan 5 elemen
array_1 = [1,3,5,7,9]
array_2 = [10,20,30,40,50]
array_3 = [5,15,25,35,45]
array_4 = [1,2,3,4,5]
```

```python
#Nomor 3.A = Perkalian pada variable array 1 dengan variable array 3
#referensi: https://datagy.io/python-subtract-two-lists/
perkalian = list()
for a, b in zip(array_1, array_3):
    hasil = a * b
    perkalian.append(hasil)
```

```python
print(perkalian)
```

```python
#Nomor 3.B = Penjumlahan pada variable array 2 dengan variable array 4
#referensi: https://datagy.io/python-subtract-two-lists/
penjumlahan = list()
for a, b in zip(array_2, array_4):
    hasil = a + b
    penjumlahan.append(hasil)
```

```python
print(penjumlahan)
```

```python
#Nomor 3.C = Jika ada irisan, dapatkan irisan pada variable 1 dengan 4
#Jawaban
print("3_C. Irisan Array 1 dan 4 =", list(set(array_1).intersection(array_4)))
```

```python
#Nomor 3.D = Ambil nilai pada urutan ke 1, 3, dan 5 pada variabel kedua
#Jawaban
print("3_D. Nilai Urutan ke 1,3,5 variabel array_2 =", [array_2[0], array_2[2], array_2[4]])
```

**NOMOR 4**

```python
#Nilai 4 siswa di 4 Kelas berbeda
nilai_kelas = {
    'kelas_MIPA1': [85, 92, 78, 88],
    'kelas_MIPA8': [95, 75, 85, 82],
    'kelas_IPS1': [85, 84, 82, 75],
    'kelas_IPS3': [83, 85, 70, 77]
}
```

```python
#Pengurangan nilai untuk siswa yang mencontek
for kelas, nilai_siswa in nilai_kelas.items():
    for i, nilai in enumerate(nilai_siswa):
        nilai_kelas[kelas][i] -= 5
```

```python
print(nilai_kelas)
```

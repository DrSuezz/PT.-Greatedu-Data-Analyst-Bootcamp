```python
def penjualan(banyak_barang, harga_barang):
    total_harga = banyak_barang * harga_barang

# Diskon 20%
    if total_harga > 1000000:
        diskon = 0.20
# Diskon 30%
    elif total_harga > 10000000:
        diskon = 0.30
# Tidak ada diskon
    else:
        diskon = 0.00
# Total setelah diskon
    total_harga_diskon = total_harga * (1 - diskon)
    return total_harga_diskon
# imput:
harga1 = penjualan(2, 10000)
harga2 = penjualan(21, 50000)
harga3 = penjualan(200, 55000)

print(harga1)
print(harga2)
print(harga3)
```

```python
def solve(arr):
    result = []
    for num in arr:
        if num % 2 == 0:
            result.append(num * 2)  # Bilangan genap, kali 2
        else:
            result.append(num - 2)  # Bilangan ganjil, kurangi 2
    return result

#input:
hasil1 = solve([1, 2, 3, 4, 5, 6, 7, 8])
hasil2 = solve([10, 100, 75, 37, 45, 20, 19, 56])
hasil3 = solve([91, 24, 32, 16, 7, 43, 59, 12, 11, 54, 30000])

print(hasil1)
print(hasil2)
print(hasil3)
```

```python
def separate_email(email):
    try:
        username, rest = email.split('@')
        domain_parts = rest.split('.')
        if len(domain_parts) >= 2:
            domain_name = domain_parts[0]
            top_level_domain = ".".join(domain_parts[1:])
            return {
                'username': username,
                'domain_name': domain_name,
                'top_level_domain': top_level_domain
            }
    except (ValueError, IndexError):
        pass
    return None

#input:
email1 = separate_email("alice.smith@gmail.com")
email2 = separate_email("jackson_wang.1992@yahoo.com")
email3 = separate_email("emilyjones_12@outlook.co.id")

print(email1)
print(email2)
print(email3)

```

```python
# Menampilkan semua data dengan City Los Angeles dan Quantity kurang dari 90

LosAngeles_data = FoodSales[
    (FoodSales['City'] == 'Los Angeles')
    &
     (FoodSales['Quantity'] < 90)]

print(LosAngeles_data)
```

```python
#Product yang paling sering dibeli di Region West
RegionWest_data = FoodSales[FoodSales['Region'] == 'West']
Pembelian = RegionWest_data['Product'].value_counts()
Produk_Laris = Pembelian.idxmax()
print(Pembelian)
print(f"Produk paling sering dibeli:{Produk_Laris}")
```

```python
# iterasi untuk membandingkan apakah nilai transaksi pada indeks tersebut lebih besar dibandingkan dengan index sebelumnya, dimulai dari index 1 (bukan 0)
FoodSales = FoodSales.reset_index(drop=True)
# Iterasi
transaksi_sebelumnya = None

for index, row in FoodSales.iterrows():
    if transaksi_sebelumnya is not None:
        transaksi = row['TotalPrice']
        if transaksi > transaksi_sebelumnya:
            print(f"Transaksi pada indeks {index} lebih besar dari transaksi sebelumnya.")
        else:
            print(f"Transaksi pada indeks {index} tidak lebih besar dari transaksi sebelumnya.")
        transaksi_sebelumnya = transaksi
    else:
        transaksi_sebelumnya = row['TotalPrice']
```

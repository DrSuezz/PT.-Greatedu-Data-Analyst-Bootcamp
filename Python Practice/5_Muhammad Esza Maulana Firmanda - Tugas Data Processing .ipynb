1. Filter data tersebut untuk hanya mengambil category Snacks, tampilkan hanya 3 baris data teratas diurutkan berdasarkan kolom Product (gunakan loc atau iloc)

```python
# Filter
snacks_data = FoodSales.loc[FoodSales['Category'] == 'Snacks'].head(3)

# Sort
snacks_data = snacks_data.sort_values(by='Product')

print(snacks_data)
```

2. Berapa selisih penjualan Product yang paling tinggi quantity penjualannya dengan yang paling rendah quantity penjualannya di city Boston?


```python
# Filter kota Boston
boston_data = FoodSales[FoodSales['City'] == 'Boston']

# Produk dengan kuantitas penjualan tertinggi dan terendah
tertinggi = boston_data['Quantity'].max()
terendah = boston_data['Quantity'].min()

# operasi
selisih_penjualan = tertinggi - terendah
print("Selisih penjualan di kota Boston:", selisih_penjualan)
```

3. Buatlah kolom baru untuk mengelompokkan UnitPrice produk menjadi mahal jika nilainya diatas rata-rata UnitPrice dan murah jika dibawah atau sama dengan rata-rata UnitPrice


```python
# mean harga unit
mean_hargaunit = FoodSales['UnitPrice'].mean()

# kategori mahal & murah
FoodSales['Harga'] = ['mahal' if x > mean_hargaunit else 'murah' for x in FoodSales['UnitPrice']]

print(FoodSales[['Product', 'UnitPrice', 'Harga']])
```

4. Buatlah function untuk menghitung berapa revenue yang didapatkan di suatu kota


```python
list_kota = FoodSales['City'].unique()
print(list_kota)
```

```python
#Fungsi
def calculate_city_revenue(data, city_name):
    # Filter data
    city_data = data[data['City'] == city_name]
    # operasi
    total_revenue = (city_data['Quantity'] * city_data['UnitPrice']).sum()

    return total_revenue
```

```python
#Pendapatan Total Setiap Kota
for city in list_kota:
    revenue = calculate_city_revenue(FoodSales, city)
    print(f"Total revenue di {city}: ${revenue:.2f}")
```
